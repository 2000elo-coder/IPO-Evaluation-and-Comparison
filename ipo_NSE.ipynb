```python
import requests
import pandas as pd

# Set headers to mimic a browser (NSE blocks bots)
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
    "Accept": "application/json",
    "Referer": "https://www.nseindia.com/"
}

# NSE IPO API endpoint
url = "https://www.nseindia.com/api/ipo-current-issue"

past_url = "https://www.nseindia.com/api/public-past-issues"

upcoming_url = "https://www.nseindia.com/api/all-upcoming-issues?category=ipo"

# Create a session to handle cookies
session = requests.Session()
session.get("https://www.nseindia.com", headers=headers)  # Initialize session

# Fetch IPO data
response = session.get(url, headers=headers)
ipos = response.json()

# Convert to DataFrame
df = pd.DataFrame(ipos)

df = df.rename(columns={
    "symbol": "Symbol",
    "companyName": "Company Name",
    "issueStartDate": "Start Date",
    "issueEndDate": "End Date",
    "issuePrice": "Price Band",
    "issueSize": "Issue Size",
    "noOfsharesBid": "Shares Bid",
    "noOfSharesOffered": "Shares Offered",
    "noOfTime": "Subscription Times"
})

# Select relevant columns
cols = ["Symbol", "Company Name", "Start Date", "End Date", "Price Band", "Issue Size",
        "Shares Offered", "Shares Bid", "Subscription Times", "status"]
df = df[cols]

print(df)
```
